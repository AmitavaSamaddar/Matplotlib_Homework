```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

merged_data = pd.merge(city_data, ride_data, on = 'city', how = 'inner')

# Display the data table for preview

merged_data = merged_data[["city", "date", "fare", "ride_id", "driver_count", "type"]]

merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Creating dataframes for Urban, Suburban and Rural
urban = merged_data.loc[merged_data['type'] == 'Urban']
suburban = merged_data.loc[merged_data['type'] == 'Suburban']
rural = merged_data.loc[merged_data['type'] == 'Rural']

#Group by Urban data by City, creating x and y axis by number of ride and avarage fare
merged_urban = urban.groupby("city")
urban_ride = merged_urban.city.count()
urban_fare   = merged_urban.fare.mean()

#Plot Urban data with driver count per city
Urban = plt.scatter(urban_ride, urban_fare, marker="o", facecolors="lightcoral", edgecolors="black",
          s=merged_urban.driver_count.mean()*10,alpha=0.75)

#Group by Suburban data by City, creating x and y axis by number of ride and avarage fare
merged_suburban = suburban.groupby("city")
suburban_ride = merged_suburban.city.count()
suburban_fare   = merged_suburban.fare.mean()

#Plot Suburban data with driver count per city
Suburban = plt.scatter(suburban_ride, suburban_fare, marker="o", facecolors="lightskyblue", edgecolors="black",
           s=merged_suburban.driver_count.mean()*10,alpha=0.75)

#Group by Rural data by City, creating x and y axis by number of ride and avarage fare
merged_rural = rural.groupby("city")
rural_ride = merged_rural.city.count()
rural_fare   = merged_rural.fare.mean()

#Plot Rural data with driver count per city
Rural = plt.scatter(rural_ride, rural_fare, marker="o", facecolors="gold", edgecolors="black",
           s=merged_rural.driver_count.mean()*10, alpha=.75)

# Obtain the x and y grid
plt.grid()

#Label x and y axis, set title for the plot
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

#Adding text to mention driver count. Position used (42,35) to match the example 
plt.text(42,35,"Note: \nCircle size correlates with driver count per city")

#Creating legend for Type
l_name = plt.legend((Urban, Suburban, Rural),("Urban","Suburban","Rural"),fontsize='small',
                    loc="upper right",title = "City Types", numpoints=1,markerscale=0.5,scatterpoints=1)

#Setting same size for the legends 
l_name.legendHandles[0]._sizes = [25]
l_name.legendHandles[1]._sizes = [25]
l_name.legendHandles[2]._sizes = [25]

#Saving the figure
plt.savefig("images/pyber_ride_data.png", bbox_inches = "tight")

#Show the plots
plt.show()

```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_by_type = merged_data.groupby("type")

total_by_type = fare_by_type.fare.sum()

#Explode parameter
expl = (0, 0, 0.1)

#Plotting pie
plt.pie(total_by_type, explode = expl, labels=total_by_type.index, colors=["gold", "lightskyblue","lightcoral"],
        autopct="%1.1f%%", shadow=True, startangle=150, 
        wedgeprops = {'linewidth' : 1,
                      'edgecolor' : "grey"})

#Add title
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("images/fares_by_citytype.png")

#Show the plot
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_by_type = fare_by_type.ride_id.count()

#Explode parameter
expl = (0, 0, 0.1)

#Plotting pie
plt.pie(ride_by_type, explode = expl, labels=total_by_type.index, colors=["gold", "lightskyblue","lightcoral"],
        autopct="%1.1f%%", shadow=True, startangle=150)

#Add title
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("images/rides_by_citytype.png")

#Show the plot
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_by_type   = city_data.groupby("type")

driver_by_type = driver_by_type.driver_count.sum()

#Explode parameter
expl = (0, 0, 0.1)

#Plotting pie
plt.pie(driver_by_type, explode = expl, labels=total_by_type.index, colors=["gold", "lightskyblue","lightcoral"],
        autopct="%1.1f%%", shadow=True, startangle=160)

#Add title
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("images/drivers_by_citytype.png")

#Show the plot
plt.show()
```
